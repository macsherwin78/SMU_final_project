```python
# import dependencies 
import pandas as pd
```

```python
# create initial dataframe of Airbnb-provided data 
df = pd.read_csv("listings.csv")
```

```python
# remove any listings with $0 price and create new dataframe 
new = df[df.price != 0]
```

```python
# describe new dataframe and verify minimum price is not $0
new.describe()
```

```python
# view new dataframe header 
new.head()
```

```python
# save new dataframe to new csv document 
new.to_csv(r'clean_listings.csv')
```
